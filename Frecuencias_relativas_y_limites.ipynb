```python
import numpy as np
import random
import matplotlib.pyplot as plt
get_ipython().magic(u'matplotlib inline')
```

```python
"""
Calculo de frecuencias relativas y absolutas para el experimento de enviar
un paquete de Medellin a Bucaramanga
"""
```

```python
#Realizacion del experimento y resultados
def experimento(N,cota):
    
    A = []                                         #Almacenamiento de resultados del experimento
    F = []                                         #Frecuencias relativas por cada N
    L = []                                         #Veces que se realiza el experimento
    
    #Experimento
    for i in xrange(0,N):
        random.seed()
        a = np.random.random()
        if a <= cota: A.append(1.0)                #1 significa si
        else: A.append(0.0)                        #0 significa no
    
    #Calculo de las frecuencias
    for i in xrange(0,len(A)):
        frec = sum(A[0:i+1])/(i*1.0 + 1.0)         #Def de frec relativa
        F.append(frec)
        L.append(i*1.0 + 1.0)
    
    #Grafico 
    plt.figure(figsize=(16,8))
    plt.plot(L,F,label='frec_lim = %.4f' %F[-1])   #Se toma el ultimo elemento de F como frec lim
    plt.title('N vs Frec_relativa')
    plt.xlabel('N')
    plt.ylabel('Frec_relativa')
    plt.ylim(0.0,1)
    plt.legend(loc='center right')
    
    return A,F,L
```

```python
r1 = experimento(10000,0.5)
```

```python
r2 = experimento(10000,0.8)
```
